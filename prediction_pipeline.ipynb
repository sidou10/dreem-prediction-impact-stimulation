## Feature engineering

###  DWT

Nous avons d'abord effectué une phase de *feature engineering* afin de réduire la dimension du problème. L'idée était d'extraire des caractéristiques des différents signaux (EEG, respiration). Nous avons utilisé pour cela la théorie des transformées en ondelettes (discrètes ici, DWT pour Discrete Wavelet Transform). A partir d'une ondelette mère que l'on translate et dilate (pour donner de nouvelles ondelettes), on représente le signal initial comme une combinaison de ces ondelettes. L'avantage de l'utilisation des ondelettes par rapport à la transformée de fourier est que la base de décomposition n'est pas forcément sinusoïdal .

Il est souvent difficile de choisir la bonne ondelette mère pour décomposer un signal donnée afin de détecter des changements caractéristiques. Les auteurs de l'article <a href = "#ref1"> [1] </a> ont résumé les différents articles de recherche s'intéresssant à l'extraction de features à partir de signaux EEG à l'aide de la DWT. L'ondelette **Daubechies 4** ou **db4** est la plus souvent utilisée. Ci-après le signal d'une telle ondelette:


*[Source](http://wavelets.pybytes.com/wavelet/db4/)*

Ainsi, nous nous sommes intéressés à l'extraction des features des différents signaux EEG à l'aide de l'ondelette db4. Les signaux de 8 sec ont été séparés en 8 segments d'une seconde puis nous avons effectué la DWT sur chacun de ces segments en utilisant 4 niveaux de décomposition (auxquels s'ajoutent le signal approximé). Nous obtenons donc 5 signaux sur chacun des 8 segments. Nous calculons alors les variables suivantes (cf. <a href = "#ref2"> [2] </a>) sur chacun des segments:
- Moyenne des valeurs absolues des coefficients
- Somme des valeurs absolue des coefficients
- Ecart-type des coefficients

Cela permet donc de réduire considérablement le nombre de varibales de type EEG de 2000 à $5 \times 8 \times 3 = 120$.

####  EEG features 

````python
# Create new variables using DWT 
signal = "eeg"
feature_types = ["mean_abs", "sum_abs", "std"]
n_segments = 8
wavelet = "db4"
level = 4
keep_approximation = True

if keep_approximation:
    feature_names = ["{}_segment_{}_level_{}_{}".format(signal, i+1, j, feature_type) for i in range(n_segments)
                for feature_type in feature_types for j in range(level+1)]
else:
    feature_names = ["{}_segment_{}_level_{}_{}".format(signal, i+1, j, feature_type) for i in range(n_segments)
                for feature_type in feature_types for j in range(1, level+1)]
    
print("Number of {} features to create: {}.".format(signal, len(feature_names)))

# Signal names 
features = [var for var in variables if signal in var]
````

````python
# Create the eeg new features
data_features = data_basic_features
data_features[feature_names] = data.loc[:, features].apply(lambda row: 
      compute_wavelets_mean_coeffs(row, feature_types=feature_types, 
                                   n_segments=n_segments, wavelet=wavelet,  
                                   level=level, 
                                   keep_approximation=keep_approximation), 
      axis=1)
````

````python
data_features.head()
````

## Gradient Boosting 

Le *Gradient Boosting* est une méthode ensembliste qui utilise la technique du boosting (regrouper plusieurs *weak learners*). A partir d'un faible modèle $F_{m}$, l'algorithme va rajouter un nouvel estimateur $h$ de manière à minimiser le résidu $y - F_{m}$, y étant la variable à prédire. Or, un tel résidu n'est autre que le gradient (négatif) de la *squared error loss function*. Donc le *Gradient Boosting* est un algorithme de déscente par gradient. Ici, les *learners* utilisés sont des arbres.

Ici, nous utilisons l'algorithme de la librairie *scikit-learn* de Python. L'algorithme a notamment les paramètres suivants que nous essaierons d'optimiser par la suite à l'aide d'un *grid-search* avec *cross-validation*:
- **min_samples_split**: nombre minimum d'observations pour qu'un noeud soit considéré pour un split. De trop grandes valeurs peuvent entraîner un *under-fitting*, de trop petites un *over-fitting*
- **min_samples_leaf**: nombre minimum d'observations dans un noeud terminal. De même, une très petite valeur peut entraîner un *over-fitting*
- **max_depth**: profondeur maximale de l'arbre. Une grande valeur pour ce paramètre peut entraîner un *over-fitting*
- **max_features**: nombre de features utilisées lors de la recherche du meilleur split (heuréstiquement égal à la racine du nombre de features). De grandes valeurs peuvent entraîner un *over-fitting*
- **learning_rate**: détermine l'impact de chaque nouvel arbre rajouté. De petites valeurs permettent une bonne généralisation mais nécessitent un plus grand nombre d'arbres
- **n_estimators**: nombre d'arbres. Une grande valeur peut entraîner un *over-fitting*
- **subsample**: nombre d'observations servant pour le training set de chaque arbre, elles sont sélectionnées de manière aléatoire

# Réseaux de neurones

Les réseaux de neurones sont un type de modèles initialement inspiré du fonctionnement des neurones biologiques. Un réseau de neurones met bout à bout des couches de neurones composés de "processeurs élémentaires" (les neurones). Chaque neurone calcule une sortie en fonction de l'information qu'il reçoit.

Ainsi, la régression logistique peut être vu comme l'exemple le plus simple de réseaux de neurones, à une couche et un neurone. Mais, la multiplication des neurones et des couches est rendu nécessaire pour résoudre des problèmes non-linéaires. Par exemple, un problème aussi simple que le XOR ne peut pas être appris efficacement avec une régression logistique.

L'idée d'utiliser des réseaux de neurones avait un objectif double :
- En pratique, c'est un type de modèle très puissant qui génère des très bon résultats
- Un réseau de neurone apprend les variables pertinentes à utiliser (couches intermédiaires) ; c'est utile pour combler notre manque de connaissances en neurosciences et en traitement du signal. L'étape de *Feature engineering* devient de l'*Architecture Engineering*

Notre travail s'est décomposé de la manière suivante :
1. Modélisation à l'aide d'un *Multi-layer Perceptron*. Les résultats ont mis en avant qu'il était nécessaire de traiter différemment les variables EEG et Respiration de par leur nature (signaux)
2. Modélisation indépendante de chacun des 2 signaux à l'aide de réseaux de neurones convolutionnels. Les résultats nous ont mené à écarter les signaux de respiration qui n'apportaient pas d'information.
3. Modèle final non séquentiel, en utilisant divers types d'inputs : le signal EEG, l'ID des utilisateurs et les métadonnées

Nous tentons dans cette partie d'entraîner un modèle de type MLP. Dans ce type de modèle, tous les neurones de deux couches successives sont connectées. Le schéma suivant montre un exemple de deux couches de trois neurones. 

*Notations :*

- $\mathbf{y}^n = \mathbf{x}^{n+1}$ : couche n+1
- $\mathbf{A}^n$ : matrice des poids de la couche n à n+1
- $\mathbf{b}^n$ : biais
- $f(.)$ : fonction d'activation (ici ReLU)




````python
# Train/cross-validation set split
X = scaled_eeg_data.join(dummies_user_data).join(scaled_meta_data).join(scaled_resp_data)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
````

### Modélisation : MLP sur toutes les variables

L'architecture du modèle suivant (outre la couche des données initiales) est :
- Deux couches cachées (hidden layers) :
    - Une couche totalement connectée (Dense) avec XXX neurones
    - Une couche "Dropout" (expliquée ci-dessous)
- Une couche final

La couche "Dropout" permet de réduire le phénomène de sur-apprentissage. La manière avec laquelle cela est fait est la suivante : à chaque *batch*, une partie aléatoire des neurones est désactivée (ici 50% des neurones, passé en paramètre de la couche). La *forward-propagation* et la *back-propagation* se fait alors sans prendre en compte ces neurones. Les poids de la matrice de passage sont mis à jour. Intuitivement, cela permet de faire une moyenne des différents modèles avec les neurones particuliés générés à chaque *batch"*.

````python
# Define architecture
def mlp(input_dim):
    model = Sequential()
    model.add(Dense(units=8, activation="relu", input_dim=input_dim))
    model.add(Dropout(0.5))
    model.add(Dense(units=1))
    model.compile(loss="mean_squared_error", optimizer="adam")
    return model
````

````python
# Train the model
input_dim = X_train.shape[1]
model = mlp(input_dim)

# Show details of the model, especially # of parameters
model.summary()
````

````python
history = model.fit(X_train, y_train, validation_data = (X_test, y_test), epochs=100, batch_size=128, verbose=0)
````

````python
# Summarize history for loss
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.title("model loss")
plt.ylabel("loss")
plt.xlabel("epoch")
plt.ylim((0.3,0.6))
plt.legend(["train", "test"], loc="upper right")
plt.show()
````

> Le MSE sur l'ensemble de cross-validation stagne autour de 0.36 après 80 epochs.

### Conclusion MLP
Le MLP fournit des résultats encourageants (RMSE ~ 0.60). Cependant, il se caractérise par une explosion très rapide du nombre des paramètres pouvant mener à du sur-apprentissage. De plus, il ne permet pas de prendre en compte le fait qu'une partie de nos données soient des signaux. Nous nous orientons donc dans la prochaine partie à une approche de type réseau de neurone convolutionnel.

## Modèle multi-input 

Dans ce modèle, nous traitons séparément les différents types d'input (4 dernières secondes du signal EEG, user_id et meta-données), et les recombinons par la suite. Notre choix de ne considérer que les 4 dernières secondes découle de l'analyse de l'importance des variables dans le modèle de Gradient Boosting en première partie.

````python
n_points_eeg = 1000
# Train set
eeg_train = np.array(X_train[eeg_col]).reshape(X_train[eeg_col].shape + (1,))[:,n_points_eeg:]
user_train = np.array(X_train[[col for col in X_train.columns if "user" in col]])
meta_train = X_train[metadata_col]

# Cross-validation set
eeg_test = np.array(X_test[eeg_col]).reshape(X_test[eeg_col].shape + (1,))[:,n_points_eeg:]
user_test = np.array(X_test[[col for col in X_test.columns if "user" in col]])
meta_test = X_test[metadata_col]
````

````python
def multi_input_model():    
    # EEG Input part
    eeg = Input(shape=(2000-n_points_eeg,1),name='eeg')
    eeg_conv1 = Conv1D(filters=4,kernel_size=5,activation='relu')(eeg)
    eeg_pool1 = AveragePooling1D(25)(eeg_conv1)
    eeg_flat = Flatten()(eeg_pool1)
    
    # User Id Input part
    user = Input(shape=(143,),name='user')
    dense_user = Dense(143, activation='relu')(user)
    dropout_user = Dropout(0.5)(dense_user)
    
    # Metadata Input part
    meta = Input(shape=(3,),name='meta')
    dense_meta = Dense(3, activation='relu')(meta)
    dropout_meta = Dropout(0.1)(dense_meta)

    # Recombined part
    merged_input = concatenate([eeg_flat, dropout_user, dropout_meta])
    dense1 = Dense(units=8, activation='relu', name='dense1')(merged_input)
    predictions = Dense(1, activation='relu')(dense1)

    model = Model(inputs=[eeg, user, meta], outputs=predictions)
    model.compile(optimizer='adam', loss='mse')
    return model
````

````python
model = multi_input_model()
model.summary()
````

````python
history = model.fit([eeg_train, user_train, meta_train], y_train, 
                    validation_data=([eeg_test, user_test, meta_test], y_test), 
                    epochs=25, batch_size=128, verbose=1)
````

````python
# Summarize history for loss
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.title("model loss")
plt.ylabel("loss")
plt.xlabel("epoch")
plt.ylim((0.2,0.4))
plt.legend(["train", "test"], loc="upper right")
plt.show()
````

> La prédiction sur le cross-validation set stagne après 8 epochs autour de MSE ~ 0.28. On choisira ce nombre d'epochs pour la soumission finale.

### Conclusion sur le modèle Multi-Input
En combinant des réseaux de neurones convolutionnels sur le signal EEG, avec des réseaux denses sur les variables classiques (méta-données et user_id), on obtient nos meilleurs résultats.
